```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

# Get all csv files in current directory
csv_files = [f for f in os.listdir('.') if f.endswith('.csv')]

# Calculate the number of rows and columns for subplots
num_files = len(csv_files)
cols = 2
rows = num_files // cols
rows += num_files % cols

# Create a position for each subplot
position = range(1, num_files + 1)

fig = plt.figure(figsize=(20, 20))

# Columns' names
columns = ['X', 'Y', 'Z']

for i, file in enumerate(csv_files):
    df = pd.read_csv(file, names=columns)
    # Create time column based on data collection frequency
    df['Time'] = pd.Series(np.arange(0, len(df)*0.02, 0.02))
    ax = fig.add_subplot(rows, cols, position[i])
    ax.plot(df['Time'], df['Y'], color='blue', linewidth=1.5)
    ax.set_title(f'Y column of {file}', fontsize=14, fontweight='bold')
    ax.set_xlabel('Time (s)', fontsize=12)
    ax.set_ylabel('Y', fontsize=12)

    # Add minor ticks and grids
    ax.minorticks_on()
    ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
    ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

    # Formatting the ticks
    ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
    ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

plt.tight_layout()
plt.show()
```
